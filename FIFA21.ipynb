```python
def curr_transf(x):
        x = x.lstrip('€')
        if not x[-1].isnumeric():
            if x[-1]== 'M':
                return float(x.rstrip('M'))*1000000
            elif x[-1]=='K':
                return float(x.rstrip('K'))*1000


        else:
            return float(x)
```

```python
def split_sum(x):
    if '+' in x:
        x= list(x.split("+"))
        b= int(x[0])+int(x[1])
        return b
```

```python
def parse_ht(ht):
    ht_strip=ht.rstrip('"')
    ht_ = ht_strip.split('\'')
    ft_in=float(ht_[0])*12
    final=float(ht_[1])+ft_in
    return final*2.54
```
